```python
#pip install missingno (Yo no lo instalo porque ya lo tengo)

#importamos los paquetes
import missingno as msno
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/Proyectos/Missingno ejemplo/all_games.csv')
proporcion = 0.1
# Eliminamos registros adrede para el ejemplo
df['user_review'] = df['user_review'].apply(lambda x: np.nan if np.random.rand() < proporcion else x)
df['release_date'] = df['release_date'].apply(lambda x: np.nan if np.random.rand() < proporcion else x)
```

```python
#Creamos una matriz de valores faltantes
msno.matrix(df)
plt.title('Matriz de valores faltantes', size=20) #Puedes modificarlo usando matplotlib
plt.show()
```

```python
msno.bar(df)
plt.title('Recuento de valores faltantes', size=20) #Puedes modificarlo usando matplotlib
plt.show()
```

```python
msno.heatmap(df)
plt.title('Correlación entre columnas con valores faltantes', size=20) #Puedes modificarlo usando matplotlib
plt.show()
```

```python
msno.dendrogram(df)
plt.title('Relación jerárquica entre columnas con valores faltantes', size=20)
plt.show()
```
